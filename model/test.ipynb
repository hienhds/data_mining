```python
from sentence_transformers import SentenceTransformer, util
import json

# Load model
model = SentenceTransformer("all-MiniLM-L6-v2")

# Load JSON d·ªØ li·ªáu (d√°n ƒëo·∫°n JSON b·∫°n g·ª≠i v√†o file `input.json`)
with open("input.json", "r") as f:
    data = json.load(f)

job = data["job"]
candidates = data["candidates"]

# 1. T·∫°o embedding cho job
job_text = job["title"] + " " + " ".join(job["skills"]) + " " + job["domain"]
job_embed = model.encode(job_text, normalize_embeddings=True)

# 2. T·∫°o embedding cho t·ª´ng candidate
candidate_scores = []
for c in candidates:
    cand_text = c["job_title"] + " " + " ".join(c["skills"]) + " " + " ".join(c["experiences"])
    cand_embed = model.encode(cand_text, normalize_embeddings=True)
    score = float(util.dot_score(job_embed, cand_embed))
    candidate_scores.append((c["id"], score))

# 3. In ra top-k ·ª©ng vi√™n ph√π h·ª£p
top_k = 5
sorted_candidates = sorted(candidate_scores, key=lambda x: -x[1])
print(f"üîç Top {top_k} ·ª©ng vi√™n ph√π h·ª£p v·ªõi job '{job['title']}':\n")
for uid, score in sorted_candidates[:top_k]:
    print(f"‚úÖ {uid} - Score: {score:.4f}")
```
